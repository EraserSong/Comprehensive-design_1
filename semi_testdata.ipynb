### 2.3 고유값 확인

```python
# 2-2. 전처리 과정 - 고유값 확인
# 특성 추출에 큰 의미를 가지지 못할 것으로 예측되는 특성은 삭제하도록 함

import pandas as pd

# 각 열에서 가장 빈번한 값의 비율을 계산하고, 그 비율이 85% 이상인 열을 찾기
threshold = 0.85  # 비율 기준값 설정 (85%)
columns_85 = []  # 삭제할 열의 목록을 저장할 리스트


for col in df.columns:
    if col == 'label':  # 'Pass_Fail' 열은 건너뛰기
        continue
    top_freq = df[col].value_counts(normalize=True).iloc[0]  # 가장 빈번한 값의 비율 계산
    if top_freq >= threshold:
        print(f'열 "{col}"는 대부분 동일한 값({top_freq*100:.2f}%)을 가집니다.')
        columns_85.append(col)
        
print(f"고유값이 85% 이상인 열의 수: {len(columns_85)}")
```

```python
# 삭제할 unique value 열 확인 
columns_85
```

```python
# columns_85 안에 들어있는 열과 동일한 이름을 가진 열을 df_filled에서 삭제하는 코드
for col in columns_85:
    del df_filled[col]

print(df_filled)
```

## 3. 상관관계 분석
### 3.1 전체 데이터의 상관관계 분석

```python
# 3. 상관관계 분석을 통한 주요 요인 확인
# 전체 데이터(센서 + 양/불) 상관관계 분석 
# 이때, label 데이터는 빼고 상관관계 분석을 진행해봄

import seaborn as sns
import matplotlib.pyplot as plt

# label 열을 제외한 나머지 열 선택
features = df_filled.drop(columns=['label'])

# 상관관계 행렬 계산
correlation_matrix = features.corr()

# 대각선을 기준으로 아래쪽 부분에 대한 마스크 생성
mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))

# 상관관계 히트맵 그리기
plt.figure(figsize=(16, 12))
sns.heatmap(correlation_matrix, annot=False, cmap="coolwarm", vmax=1, vmin=-1, mask=mask)
plt.title('Correlation Heatmap (Total data)')
plt.show()
```

```python
correlation_matrix
```

### 각 요인 군집별(A,B,...,O) 상관관계 확인

원 데이터 셋에는 A,B,C,...,O 등 다양한 요인(?)이 존재한다. 동일 알파벳으로 묶이는 요인끼리는 서로의 관계성이 높을 것이라 보고 동일 군집끼리 상관관계를 분석해봄으로써, 주요인 확인에 다가가보고자 한다.

```python
# 알파벳으로 시작하는 열들을 그룹화하여 상관관계 분석
alphabet_groups = {}

for col in df_filled.columns:
    first_letter = col[0]
    if first_letter not in alphabet_groups:
        alphabet_groups[first_letter] = []
    alphabet_groups[first_letter].append(col)

# 각 그룹별 상관관계 히트맵 시각화
fig, axes = plt.subplots(nrows=len(alphabet_groups), ncols=1, figsize=(16, 4 * len(alphabet_groups)))

for i, (first_letter, columns) in enumerate(alphabet_groups.items()):
    features = df_filled[columns]
    correlation_matrix = features.corr()
    mask = np.triu(np.ones_like(correlation_matrix, dtype=bool))
    sns.heatmap(correlation_matrix, annot=False, cmap="coolwarm", vmax=1, vmin=-1, mask=mask, ax=axes[i])
    axes[i].set_title(f'Correlation Heatmap for Columns starting with "{first_letter}"')

plt.tight_layout()
plt.show()
```

### 결과
예상했던 것 보다는 크게 관계성이 뚜렸하다 보기는 어려우나, 전체 상관관계 분석과 비교 했을때, 주요 관계요인을 좀 더 잘 확인할 수 있다.

## 4. 차원 축소

### 4.1 전체 데이터 T-sne

```python
# 특성 열과 라벨 열 분리

X = df_filled.drop(columns=['label'])  # 특성 열
y = df_filled['label']  # 라벨 열

print(X)
print(y)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # 3차원 플로팅을 위한 모듈
from sklearn.preprocessing import StandardScaler, MinMaxScaler
from sklearn.manifold import TSNE

# T-SNE 모델 초기화
tsne = TSNE(n_components=3, random_state=42)

fig = plt.figure(figsize=(24, 16))

# 시각화할 데이터와 스케일러 목록
datasets = [(X, "Original Data"), (StandardScaler().fit_transform(X), "Standard Scaled Data"), 
            (MinMaxScaler().fit_transform(X), "MinMax Scaled Data")]

for i, (X_scaled, title) in enumerate(datasets, start=1):
    # T-SNE를 사용하여 3차원으로 데이터 축소
    X_tsne = tsne.fit_transform(X_scaled)

    # subplot 추가 (전체 데이터)
    ax = fig.add_subplot(3, 4, i*4-3, projection='3d')
    scatter = ax.scatter(X_tsne[:, 0], X_tsne[:, 1], X_tsne[:, 2], c=y, cmap='viridis')
    ax.set_title(f'T-SNE Visualization ({title})')
    ax.set_xlabel('T-SNE Dimension 1')
    ax.set_ylabel('T-SNE Dimension 2')
    ax.set_zlabel('T-SNE Dimension 3')
    plt.colorbar(scatter, ax=ax, label='label')

    # subplot 추가 (y가 0인 데이터)
    ax = fig.add_subplot(3, 4, i*4-2, projection='3d')
    scatter = ax.scatter(X_tsne[y == 0, 0], X_tsne[y == 0, 1], X_tsne[y == 0, 2], c='blue', cmap='viridis')
    ax.set_title(f'T-SNE Visualization ({title}) - Label 0')
    ax.set_xlabel('T-SNE Dimension 1')
    ax.set_ylabel('T-SNE Dimension 2')
    ax.set_zlabel('T-SNE Dimension 3')

    # subplot 추가 (y가 1인 데이터)
    ax = fig.add_subplot(3, 4, i*4-1, projection='3d')
    scatter = ax.scatter(X_tsne[y == 1, 0], X_tsne[y == 1, 1], X_tsne[y == 1, 2], c='red', cmap='viridis')
    ax.set_title(f'T-SNE Visualization ({title}) - Label 1')
    ax.set_xlabel('T-SNE Dimension 1')
    ax.set_ylabel('T-SNE Dimension 2')
    ax.set_zlabel('T-SNE Dimension 3')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler, MinMaxScaler

# 스케일러 목록
scalers = [("No Scaling", None), ("Standard Scaler", StandardScaler()), ("MinMax Scaler", MinMaxScaler())]

# 각 스케일러에 대해 최적의 차원 수를 확인하고 시각화
for scaler_name, scaler in scalers:
    # PCA 모델 초기화
    pca = PCA()
    
    # 스케일링 적용
    if scaler is not None:
        X_scaled = scaler.fit_transform(X)
    else:
        X_scaled = X
    
    # PCA 적용하여 설명된 분산 확인
    pca.fit(X_scaled)
    
    # 설명된 분산 비율의 누적 합 계산
    explained_variance_ratio_cumsum = pca.explained_variance_ratio_.cumsum()
    
    # 최적의 차원 수 결정
    optimal_dimension = next(i+1 for i, explained_variance in enumerate(explained_variance_ratio_cumsum) if explained_variance >= 0.95)
    
    # 시각화
    plt.plot(range(1, len(explained_variance_ratio_cumsum) + 1), explained_variance_ratio_cumsum, marker='o', label=scaler_name)
    plt.xlabel('Number of Components')
    plt.ylabel('Cumulative Explained Variance Ratio')
    plt.title('Cumulative Explained Variance Ratio vs Number of Components')
    plt.axvline(optimal_dimension, color='red', linestyle='--', label=f'Optimal Dimension: {optimal_dimension}')
    plt.legend()
    plt.grid(True)
    plt.tight_layout()

# 그래프 출력
plt.show()
```

## K-fold 교차 검증
연구에 사용된 데이터가 매우 부족한 상황으로, 정확도를 높이기 위해서 
1) 전체 데이터셋, 즉 앞서 만들어진 각각의 불량 데이터셋에 K개의 Train/Test셋을 만든다

2) 각 fold의 Train데이터에 SMOTE를 진행하여 데이터 불균형을 해결한다

    이때, Train 데이터에만 SMOTE를 진행하는 이유는, test데이터에도 SMOTE가 적용되었을때 과적합의 우려가 있기 때문이다.
    
3) 만들어진 각 불량 군집별 데이터셋의 최종 K-fold에서 각 fold에 NN,RF,LR을 모델을 훈련

4) 각 데이터 셋의 fold에 모델별로 예측 진행해서 정확도/예측값 확인

### 1) 각 불량 데이터 셋 별 K-fold 진행

```python
from sklearn.model_selection import KFold

# 데이터셋 목록
datasets = [D_resampled1, D_resampled2, D_resampled3, D_resampled4]  # 각 데이터셋에 대한 리스트로 가정

# K-fold 교차 검증을 위한 K 값
k = 3

# 데이터셋별로 K-fold 교차 검증을 수행
for i, dataset in enumerate(datasets, start=1):
    print(f"Dataset {i}")
    kf = KFold(n_splits=k)
    fold = 1
    for train_index, test_index in kf.split(dataset):
        train_data, test_data = dataset.iloc[train_index], dataset.iloc[test_index]
        
        # 정상 데이터와 불량 데이터 개수 세기
        normal_train_count = (train_data['label'] == 0).sum()
        faulty_train_count = (train_data['label'] == 1).sum()
        normal_test_count = (test_data['label'] == 0).sum()
        faulty_test_count = (test_data['label'] == 1).sum()
        
        # 각 폴드의 정상 데이터와 불량 데이터 비율 출력
        print(f"Fold {fold}:")
        print(f"Train - Normal: {normal_train_count}, Faulty: {faulty_train_count}, Ratio: {normal_train_count / len(train_data):.2f}:{faulty_train_count / len(train_data):.2f}")
        print(f"Test - Normal: {normal_test_count}, Faulty: {faulty_test_count}, Ratio: {normal_test_count / len(test_data):.2f}:{faulty_test_count / len(test_data):.2f}")
        fold += 1
    print("\n")
```

이러한 문제점을 해결하기 위해 기존의 K-fold 교차검증 방법이 아닌, Stratified K-fold라는 방법을 이용하여 해결하였다. 
Stratified K-fold란, 불균형한 분포도를 가진 레이블 데이터 집하을 위한 K-fold방식으로, 원본 데이터의 레이블 분포를 먼저 고려한 뒤 이 분포와 동일하게 Train/Test 세트를 분배하는 방법이다.

### 2) 각 fold의 Train data에 SMOTE 적용

그런 다음 이어서 만들어진 각 폴드 속 train data에 SMOTE를 적용하여 데이터 불균형을 해결하도록 한다.

```python
from sklearn.model_selection import StratifiedKFold
from imblearn.over_sampling import SMOTE

# 데이터셋 목록
datasets = [D_resampled1, D_resampled2, D_resampled3, D_resampled4]  # 각 데이터셋에 대한 리스트로 가정

# K-fold 교차 검증을 위한 K 값
k = 3

# 데이터셋별로 Stratified K-fold 교차 검증을 수행
for i, dataset in enumerate(datasets, start=1):
    print(f"Dataset {i}")
    skf = StratifiedKFold(n_splits=k)
    fold = 1
    for train_index, test_index in skf.split(dataset, dataset['label']):
        train_data, test_data = dataset.iloc[train_index], dataset.iloc[test_index]
        
        # 정상 데이터와 불량 데이터 개수 세기
        normal_train_count = (train_data['label'] == 0).sum()
        faulty_train_count = (train_data['label'] == 1).sum()
        normal_test_count = (test_data['label'] == 0).sum()
        faulty_test_count = (test_data['label'] == 1).sum()
        
        # 각 폴드의 정상 데이터와 불량 데이터 비율 출력
        print(f"Fold {fold}:")
        print(f"Train - Normal: {normal_train_count}, Faulty: {faulty_train_count}, Ratio: {normal_train_count / len(train_data):.2f}:{faulty_train_count / len(train_data):.2f}")
        print(f"Test - Normal: {normal_test_count}, Faulty: {faulty_test_count}, Ratio: {normal_test_count / len(test_data):.2f}:{faulty_test_count / len(test_data):.2f}")
        
        # SMOTE 적용
        smote = SMOTE(k_neighbors=1, random_state=42)
        X_train_resampled, y_train_resampled = smote.fit_resample(train_data.drop('label', axis=1), train_data['label'])
        
        # 새로 생성된 데이터셋에 대한 정보 출력
        print(f"Train (after SMOTE) - Normal: {len(y_train_resampled) - sum(y_train_resampled)}, Faulty: {sum(y_train_resampled)}, Ratio: {(len(y_train_resampled) - sum(y_train_resampled)) / len(y_train_resampled):.2f}:{sum(y_train_resampled) / len(y_train_resampled):.2f}")
        
        fold += 1
    print("\n")
```

### 3) 만들어진 각 불량 군집별 데이터셋의 최종 K-fold에 예측 진행

(1) Random Forest 을 이용한 모델 성능 

```python
import numpy as np
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from imblearn.over_sampling import SMOTE
from sklearn.model_selection import StratifiedKFold

# 데이터셋 목록
datasets = [D_resampled1, D_resampled2, D_resampled3, D_resampled4]  # 각 데이터셋에 대한 리스트로 가정

# K-fold 교차 검증을 위한 K 값
k = 3

# 모델 초기화
random_forest_model = RandomForestClassifier(random_state=42)

# 각 폴드에서의 정확도를 저장할 리스트
accuracies = []

# 데이터셋별로 Stratified K-fold 교차 검증을 수행
for i, dataset in enumerate(datasets, start=1):
    print(f"Dataset {i}")
    skf = StratifiedKFold(n_splits=k)
    fold = 1
    for train_index, test_index in skf.split(dataset, dataset['label']):
        train_data, test_data = dataset.iloc[train_index], dataset.iloc[test_index]
        
        # 정상 데이터와 불량 데이터 개수 세기
        normal_train_count = (train_data['label'] == 0).sum()
        faulty_train_count = (train_data['label'] == 1).sum()
        normal_test_count = (test_data['label'] == 0).sum()
        faulty_test_count = (test_data['label'] == 1).sum()
        
        # 각 폴드의 정상 데이터와 불량 데이터 비율 출력
        print(f"Fold {fold}:")
        print(f"Train - Normal: {normal_train_count}, Faulty: {faulty_train_count}, Ratio: {normal_train_count / len(train_data):.2f}:{faulty_train_count / len(train_data):.2f}")
        print(f"Test - Normal: {normal_test_count}, Faulty: {faulty_test_count}, Ratio: {normal_test_count / len(test_data):.2f}:{faulty_test_count / len(test_data):.2f}")
        
        # SMOTE 적용
        smote = SMOTE(k_neighbors=1, random_state=42)
        X_train_resampled, y_train_resampled = smote.fit_resample(train_data.drop('label', axis=1), train_data['label'])
        
        # 랜덤 포레스트 모델 훈련
        random_forest_model.fit(X_train_resampled, y_train_resampled)
        
        # 테스트 데이터로 예측
        RF_predictions = random_forest_model.predict(test_data.drop('label', axis=1))
        
        # 정확도 계산 및 출력
        accuracy = accuracy_score(test_data['label'], RF_predictions)
        accuracies.append(accuracy)
        print(f"Fold {fold} Accuracy: {accuracy:.4f}")
        
        fold += 1
    print("\n")
    
# 각 폴드에서의 정확도 평균 출력
print(f"Average Accuracy: {sum(accuracies) / len(accuracies):.2f}")
```
